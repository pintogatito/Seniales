# Propiedades de las señales senoidales

En este notebook se explicará de manera interactiva las propiedades de las señales senoidales según Comer.

Una señal senoidal es un modelo matemático que se representa de la siguiente manera: $Asin(2\pi f(t + \phi))$. Los parámetros de la función senoidal son estos, $A$ es el parámetro que indica la amplitud de la señal, $f$ es la frecuencia de la señal y $\phi$ es la fase. La variable $t$ del modelo representa el tiempo.

La siguiente imagen muestra una señal senoidal con parámetros $A=1, f=1$ y $\phi =1$,  y . La gráfica presenta la señal senoidal en el rango $[0,2\pi]$.

En las siguientes secciones de este notebook se explican cada una de las propiedades de la señal senoidal.

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

amplitude = 1.
frequency = 1.
phase = 1.
x_max = 2*pi
x_min = 0
step = 0.01

# x = np.arange(x_min, x_max, step)
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]

y = [sine_wave(amplitude, frequency, phase, time) for time in x]

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.show()
```

Una **señal periódica** tiene la propiedad de que su valor se repite luego de un desplazamiento de tiempo $T$ la cual se puede representar como: $x(t) = x(t+T)$. El valor de $T$ es conocido como el periodo de la señal.

La relación entre el periodo T y la frecuencia f se da por la ecuacion:
$$f = \frac{1}{T} \text{, en Hz}$$
$$\omega = 2 \pi$$ 
$$f = \frac{2 \pi}{T} \text{, en rad/s}$$
Otra forma de escribir la ecuación es:
$$x(t) = x(t + mT)$$
 
La señal senoidal es peridodica debido a que si tomas un conjunto $A, x \in \reals$, posteriormente podras tomar un conjunto siguiente $B, z \in \reals$ tal que, $f(x) = f(z)$, para todos los valores de los conjuntos de $x$ y $z$.

La amplitud es la distancia entre el punto más alejado de una onda y el punto de equilibrio o medio. La intensidad máxima de la señal o “valor pico” puede ser representada en valores de voltaje o corriente.

Se puede representar matemáticamente como una magnitud física $\psi$, cuyo valor varía con el tiempo y de un punto a otro del espacio, de la siguiente manera:
$$
{\displaystyle {\frac {\partial ^{2}\psi }{\partial x^{2}}}+{\frac {\partial ^{2}\psi }{\partial y^{2}}}+{\frac {\partial ^{2}\psi }{\partial z^{2}}}={\frac {1}{v_{p}^{2}}}{\frac {\partial ^{2}\psi }{\partial t^{2}}}}{\displaystyle {\frac {\partial ^{2}\psi }{\partial x^{2}}}+{\frac {\partial ^{2}\psi }{\partial y^{2}}}+{\frac {\partial ^{2}\psi }{\partial z^{2}}}={\frac {1}{v_{p}^{2}}}{\frac {\partial ^{2}\psi }{\partial t^{2}}}}
$$
Donde ${\displaystyle v_{p}\,}$ es la velocidad de propagación de la perturbación. Para una onda plana que se propaga en dirección x, la solución de la ecuación anterior es:
$$
{\displaystyle \psi (x,y,z,t)=f(x-v_{p}t)+g(x+v_{p}t)\,}{\displaystyle \psi (x,y,z,t)=f(x-v_{p}t)+g(x+v_{p}t)\,}
$$
Y en ese caso la amplitud se define como:
$$
{\displaystyle A_{\psi }=\max _{t\in \mathbb {R} }|f(x-v_{p}t)+g(x+v_{p}t)|}{\displaystyle A_{\psi }=\max _{t\in \mathbb {R} }|f(x-v_{p}t)+g(x+v_{p}t)|}
$$

```python
parametro_x = 6

from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
# if using a jupyter notebook 
%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*(t + phi))
    
amplitude = parametro_x
frequency = 1.
phase = 0.
x_max = 2*pi
x_min = 0
step = 0.01

x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [sine_wave(amplitude, frequency, phase, time) for time in x]

ax = plt.axes()
ax.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.grid(True)
ax.annotate('Amplitud',xy=(1.25,0), xytext=(1.25,amplitude/2), arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Amplitud',xy=(1.25,amplitude), xytext=(1.25,amplitude/2), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

La diferencia de fase o ángulo de desfasamiento indica el ángulo que se encuentra atrasado o adelantado respecto a otra señal.

La representacion matematica del angulo de una fase puede describirse como:
$${\displaystyle A(x,t)=A_{0}\cos(kx-\omega t+\varphi _{0})\,}$$
,donde $A$ es la amplitud y $\omega$ la pulsación,  $k$ el número de onda circular o angular, $t$ el tiempo y $x$ la posición.

El ángulo de fase de esta onda es el argumento ${\displaystyle \scriptstyle \varphi =(kx-\omega t+\varphi _{0})}$ en el caso general toda onda estacionaria puede representarse mediante una función del tipo:
$$
{\displaystyle \mathbf {A} (\mathbf {r} ,t)=\mathbf {A} _{0}(\mathbf {r} )\cos(\mathbf {k} \cdot \mathbf {r} -\omega t+\varphi _{0})}{\displaystyle \mathbf {A} (\mathbf {r} ,t)=\mathbf {A} _{0}(\mathbf {r} )\cos(\mathbf {k} \cdot \mathbf {r} -\omega t+\varphi _{0})}
$$

# Longitud de Onda

La longitud de onda representada con el simbolo $\lambda$ (lambda) se uiliza para representa la distancia que recorre una perturbación periódica que se propaga por un medio en un ciclo, es decir, es el espacio que hay entre un maximo de la función senoidal y el maximo consecutivo.

En el siguiente codigo de python, se puede modificar la variable "frequency" para poder ver que entre mas alta es la frecuencia, mas baja es la longitud de onda.

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
# if using a jupyter notebook
%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

amplitude = 1.
frequency = 1
phase = 0
x_max = 4
x_min = 0
step = 0.01

# x = np.arange(x_min, x_max, step)
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [sine_wave(amplitude, frequency, phase, time) for time in x]
plt.plot(x,y)

# variable que controla en que pico aparece la linea
pico = 3

inicio = (3/12)/frequency + ((pico/frequency)-(1/frequency))
fin = inicio + 1/frequency
plt.hlines(1,inicio,fin,color='y')
plt.text((inicio+fin)/2,0.9, 'λ')
plt.show()
```

# Señal compuesta

Una señal compuesta es una señal la cual está hecha de la suma de varias ondas senoidales simples, puede haber una infinidad de ellas formadas por diferentes frecuencias,amplitudes y fases.

En el siguiente codigo, se grafican dos ondas con amplitudes, frecuencias y fases diferentes, se puede modificar las variables de amplitud,frecuencia y fase para variarlas.

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
# if using a jupyter notebook
%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)
#distancias y paso de dibujado
x_max = 4
x_min = 0
step = 0.01

#datos de la primera onda (modificables)
amplitude = 0.3
frequency = 0.5
phase = 0

#datos de la segunda onda (modificables)
amplitude1 = 0.5
frequency1 = 0.7
phase1 = 0.5

x1 = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y1 = [sine_wave(amplitude, frequency, phase, time)  for time in x]
plt.subplot(3,1,1)
plt.plot(x1,y1)
plt.title("Onda 1")

x2 = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y2 = [sine_wave(amplitude1, frequency1, phase1, time)  for time in x]
plt.subplot(3,1,2)
plt.plot(x2,y2)
plt.title("Onda 2")

x3 = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y3 = [sine_wave(amplitude, frequency, phase, time) + sine_wave(amplitude1, frequency1, phase1, time)  for time in x]
plt.subplot(3,1,3)
plt.plot(x3,y3)
plt.title("onda compuesta")



plt.show()
```

## Importancia de las señales senoidales y compuestas

Las señales senoidales son usadas en distintos ambitos, aunque el de nuestro interes es en de las redes, mas escificamente en el envio de bits a través de la red, su uso se base en la amplitud positiva y negativa de dicha señal, ya que nos hace parecer la representación de unos y ceros. aunque una señal así por sí sola no nos sirve puesto que siempre varia de arriba a abajo infinitamente, aquí es donde entran las señales compuestas ya que con ellas podemos cambiar la forma de la señal que se está mandando.

La modulación se define como el proceso de transformar información de su forma original a una forma más adecuada para la transmisión. Cuando modulamos una señal, la estamos adaptando para poder ser transmitida a través de un canal de comunicaciones, este proceso que se realiza se hace agregando señales a la principal contrastando esta.

En el siguiente codigo se muestra como se realiza la amplitud de modulación de una onda, para realizar esto se hace uno de una señal senoidal(amplitude,frequency,phase), una señal portadora(Amplitud_P,Frecuencia_P,Fase_P) y el indice de modulación(ka).

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

#distancias y paso de dibujado
x_max = 4
x_min = 0
step = 0.01

#datos de la primera onda
amplitude = 6
frequency = 1
phase = 0

Amplitud_P= 1    #amplitud de la señal portadora
Frecuencia_P= 10  #amplitud de la señal portadora
Fase_P = 0

ka=1        #indice de la modulación

#primera onda
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y1 = [sine_wave(amplitude, frequency, phase, time) for time in x]
plt.subplot(3,1,1)
plt.plot(x,y1)
plt.title("señal senoidal")

#segunda onda
y2 = [sine_wave(Amplitud_P, Frecuencia_P, Fase_P, time) for time in x]
plt.subplot(3,1,2)
plt.plot(x,y2)
plt.title("señal portadora")

#tercera onda
y2 = [1+ka*sine_wave(amplitude, frequency, phase, time)*sine_wave(Amplitud_P, Frecuencia_P, Fase_P, time) for time in x]
plt.subplot(3,1,3)
plt.plot(x,y2)
plt.title("señal Amplitud modulada")

plt.show()
```

## Ancho de banda

El ancho de banda es una medida que indica la capacidad máxima de un enlace de comunicaciones (cableado o inalámbrico) para transmitir datos a través de una conexión de red en un determinado lapso, es decir es la velocidad de transferencia de datos

Esta medida se representa en número de bits, kilobits, megabits o gigabits que se transmiten a lo largo de un segundo.

## Transformada de Fourier
La transformada de fourier nos sirve para transformar señales entre el dominio del tiempo y el dominio de la frecuencia.

la transformada de fourier está dada por la siguiente formula:

$\displaystyle g(\xi) = {\frac{1}{\sqrt{2\pi}}} \int_{-\infty}^{+\infty} {\displaystyle f(x)e^{-i\epsilon x}dx}$

La tranaformada de fourier en el procesamiento de señales, suele considerarse como la descomposición de una señal en componentes de frecuencias diferentes, es decir, $\displaystyle g$ corresponde al espectro de frecuencias de la señal $\displaystyle f$. En la grafica siguiente se muestra como se usa la transformada de fourier para encontrar un espectro de frecuencia usando funciones de la biblioteca numpy, fft y fftfreq, estas funciones nos calcula la transformada rapida de fourier.

En el caso del seno , la transformada de fourier está dada por la siguiente definición:

$sin(\omega_0 \tau) = {{\frac{\pi}{i}} [\delta(\omega - \omega_0) - \delta(\omega - \omega_0)]}$


##  Espectro de frecuencia
El espectro de frecuencias nos muestra como una señal compuesta se descompone en sus señales senoidales basicas con diferente amplitud, frecuencia y fase. esta se obtiene con la transformada de fourier explicada en el punto anterior, a continuación se muestra el espectro de una onda senoidal con amplitud=1,frecuencia = 1 y fase = 0. Notese como el espectro queda solamente con un pico a la izquierda el cual seria una onda senoidal con amplitud 1 y ya que no está compuesta por otras señales, no hay otro pico mas adelante , en caso de haberlo estarian las ondas en las que se descompone.

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
# if using a jupyter notebook
%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

#datos de la primera onda
amplitude = 0.3
frequency = 0.5
phase = 0

#datos de la segunda onda
Amplitud_P = 1
Frecuencia_P = 7
Fase_P = 0.5

#datos de la segunda onda
amplitude2 = 1
frequency2 = 2
phase2 = 1

x_max = 2*pi
x_min = 0
step = 0.01
muestreo = 600
T = 1.0 / x_max

# x = np.arange(x_min, x_max, step)
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [sine_wave(amplitude, frequency, phase, time) + sine_wave(Amplitud_P, Frecuencia_P, Fase_P, time) + sine_wave(amplitude2, frequency2, phase2, time) for time in x]
plt.subplot(3,1,1)
plt.plot(x,y)
plt.title("Señal senoidal")

yf = np.fft.fft(y) #obteniendo la transformada rapida de Fourier
xf = np.fft.fftfreq(len(y),d= x[1] - x[0])
plt.subplot(3,1,2)
plt.plot(xf, abs(yf)**2)
plt.title("Espectro de frecuencia")
plt.show()
```
